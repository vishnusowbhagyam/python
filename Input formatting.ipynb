### Read list from user

```python
#integers
lst = list(map(int,input("enter a numbers into the list:").split()))
#string
lst1 = list(map(str,input("enter a strings into the list:").split()))
#float
lst2 = list(map(float,input("enter a float_numbers into the list:").split()))
```

```python
print(type(lst))
```

### Read tuple from user

```python
#integers
tuple1 = tuple(map(int,input("enter a numbers into the tuple:").split()))
#string
tuple2 = tuple(map(str,input("enter a strings into the tuple:").split()))
#float
tuple3 = tuple(map(float,input("enter a float_numbers into the tuple:").split()))
```

### read dictionary from user

```python
n = int(input())
d = {}
for i in range(n):
    key = int(input())
    value = int(input())
    d[key] = value
print(d)
```
