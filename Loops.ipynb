### For loop problems

```python
#Print numbers from 1 to 10
for i in range(1,11):
    print(i)
```

```python
# Print even numbers between 1 to 20
for i in range(1,21):
    if i % 2 == 0:
        print(i)
```

```python
# Find sum of first 10 natural numbers
sum = 0
for i in range(11):
    sum += i
print(sum)
```

```python
# Print a right-angled triangle
n = int(input())
for i in range(1,n+1):
    print('* '*i)
    print()
```

### While Loop

```python
# reverse a number
num = 1234
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print("Reversed Number:", rev)
```
